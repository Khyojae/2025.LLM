```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Input, Dense, Flatten, Reshape, Conv2D, Conv2DTranspose
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# [프로그램 10-2(a)] MNIST를 이용한 오토인코더 구현
# ------------------------------------------------------------------

# MNIST 데이터를 읽고 신경망에 입력할 준비
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))
x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))

zdim = 32 # 잠복 공간의 차원

# 오토인코더의 인코더 부분 설계
encoder_input = Input(shape=(28, 28, 1))
x = Conv2D(32, (3, 3), activation='relu', padding='same', strides=(1, 1))(encoder_input)
x = Conv2D(64, (3, 3), activation='relu', padding='same', strides=(2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same', strides=(2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same', strides=(1, 1))(x)
x = Flatten()(x)
encoder_output = Dense(zdim)(x)
model_encoder = Model(encoder_input, encoder_output)
print("--- Encoder Model Summary ---")
model_encoder.summary()

# 오토인코더의 디코더 부분 설계
decoder_input = Input(shape=(zdim,))
x = Dense(3136)(decoder_input) # 7*7*64 = 3136
x = Reshape((7, 7, 64))(x)
x = Conv2DTranspose(64, (3, 3), activation='relu', padding='same', strides=(1, 1))(x)
x = Conv2DTranspose(64, (3, 3), activation='relu', padding='same', strides=(2, 2))(x)
x = Conv2DTranspose(32, (3, 3), activation='relu', padding='same', strides=(2, 2))(x)
x = Conv2DTranspose(1, (3, 3), activation='relu', padding='same', strides=(1, 1))(x)
decoder_output = x
model_decoder = Model(decoder_input, decoder_output)
print("\n--- Decoder Model Summary ---")
model_decoder.summary()

# 인코더와 디코더를 결합하여 오토인코더 모델 구축
model_input = encoder_input
model_output = model_decoder(encoder_output)
model = Model(model_input, model_output)
print("\n--- Full Autoencoder Model Summary ---")
model.summary()

# 오토인코더 학습
model.compile(optimizer='Adam', loss='mse')
model.fit(x_train, x_train, epochs=5, batch_size=128, shuffle=True, validation_data=(x_test, x_test))

# 복원 실험 1: x_test를 복원하는 예측 실험
decoded_img = model.predict(x_test)

# 테스트 샘플에 대해 복원 품질 평가
n = 10
plt.figure(figsize=(20, 4))
plt.suptitle("[Program 10-2(a)] Autoencoder Reconstruction")
for i in range(n):
    # 원본 이미지
    plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.xticks([]); plt.yticks([])

    # 복원된 이미지
    plt.subplot(2, n, i + n + 1)
    plt.imshow(decoded_img[i].reshape(28, 28), cmap='gray')
    plt.xticks([]); plt.yticks([])
plt.show()


# ------------------------------------------------------------------
# [프로그램 10-2(b)] 학습된 오토인코더를 생성 모델로 활용
# ------------------------------------------------------------------

# 생성 실험 1: 첫 번째 샘플의 잠복 공간 표현에 잡음을 섞어 새로운 샘플 생성
x0 = x_test[0]
z = model_encoder.predict(x0.reshape(1, 28, 28, 1))
print("\nOriginal latent space vector (first test sample):")
print(np.round(z, 3))

zz = np.zeros((20, zdim))
for i in range(20):
    zz[i] = z[0] + (i - 10) / 10.0 # z[0]에 -1.0 ~ +0.9 잡음 추가

generated_img = model_decoder.predict(zz)

plt.figure(figsize=(20, 4))
plt.suptitle("[Program 10-2(b)] Generation Experiment 1: Adding Noise")
for i in range(20):
    plt.subplot(2, 10, i + 1)
    plt.imshow(generated_img[i].reshape(28, 28), cmap='gray')
    plt.xticks([]); plt.yticks([])
    plt.title('noise=' + str((i - 10) / 10.0))
plt.show()

# 생성 실험 2: 같은 부류의 두 샘플 사이를 보간하여 새로운 샘플 생성
# x_test[4]와 x_test[6]은 둘 다 숫자 '4'임
x4_6 = np.array([x_test[4], x_test[6]])
z = model_encoder.predict(x4_6)

zz = np.zeros((20, zdim))
for i in range(20):
    alpha = i / (20.0 - 1.0)
    zz[i] = (1.0 - alpha) * z[0] + alpha * z[1] # z[0]과 z[1] 사이를 보간

generated_img = model_decoder.predict(zz)

plt.figure(figsize=(20, 4))
plt.suptitle("[Program 10-2(b)] Generation Experiment 2: Interpolation")
for i in range(20):
    plt.subplot(2, 10, i + 1)
    plt.imshow(generated_img[i].reshape(28, 28), cmap='gray')
    plt.xticks([]); plt.yticks([])
    plt.title('alpha=' + str(round(i / (20.0 - 1.0), 3)))
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

# 1. 데이터 준비
(x_train, _), (x_test, _) = mnist.load_data()
x_train = np.expand_dims(x_train.astype('float32') / 255., -1)
x_test = np.expand_dims(x_test.astype('float32') / 255., -1)

latent_dim = 2

# 2. Sampling Layer
class Sampling(layers.Layer):
    """Reparameterization trick을 사용하여 (mu, log_var)로부터 z를 샘플링합니다."""
    def call(self, inputs):
        mu, log_var = inputs
        eps = tf.random.normal(shape=tf.shape(mu))
        return mu + tf.exp(0.5 * log_var) * eps

# 3. 인코더 정의
def build_encoder(latent_dim):
    inputs = layers.Input(shape=(28, 28, 1))
    x = layers.Flatten()(inputs)
    x = layers.Dense(256, activation='relu')(x)
    mu = layers.Dense(latent_dim, name="mu")(x)
    log_var = layers.Dense(latent_dim, name="log_var")(x)
    z = Sampling()([mu, log_var])
    return Model(inputs, [z, mu, log_var], name="encoder")

# 4. 디코더 정의
def build_decoder(latent_dim):
    decoder_inputs = layers.Input(shape=(latent_dim,))
    d = layers.Dense(256, activation='relu')(decoder_inputs)
    d = layers.Dense(28 * 28, activation='sigmoid')(d)
    outputs_decoder = layers.Reshape((28, 28, 1))(d)
    return Model(decoder_inputs, outputs_decoder, name="decoder")

# 5. VAE 모델 (test_step 추가)
class VAE(Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        # 손실을 추적하기 위한 지표(metric)
        self.total_loss_tracker = tf.keras.metrics.Mean(name="total_loss")
        self.recon_loss_tracker = tf.keras.metrics.Mean(name="recon_loss")
        self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.recon_loss_tracker,
            self.kl_loss_tracker,
        ]

    def call(self, inputs):
        z, _, _ = self.encoder(inputs)
        reconstruction = self.decoder(z)
        return reconstruction

    def train_step(self, data):
        # data는 (x, y) 형태지만, fit(x, ...)로 호출하면 data는 x가 됩니다.
        x = data
        if isinstance(data, tuple):
             x = data[0]

        with tf.GradientTape() as tape:
            z, mu, log_var = self.encoder(x)
            reconstruction = self.decoder(z)

            # 손실 계산
            recon_loss = tf.reduce_mean(
                tf.reduce_sum(
                    tf.keras.losses.binary_crossentropy(x, reconstruction), axis=(1, 2)
                )
            )
            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(
                    1 + log_var - tf.square(mu) - tf.exp(log_var), axis=1
                )
            )
            total_loss = recon_loss + kl_loss

        # 그래디언트 계산 및 가중치 업데이트
        grads = tape.gradient(total_loss, self.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))

        # 지표 업데이트
        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        return {
            "loss": self.total_loss_tracker.result(),
            "recon_loss": self.recon_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

    # *** ⭐️ 새로 추가된 부분 ⭐️ ***
    def test_step(self, data):
        # data는 (x, y) 형태지만, validation_data=x_test로 전달하면 data는 x_test가 됩니다.
        x = data
        if isinstance(data, tuple):
             x = data[0]

        # 추론 (GradientTape 없음)
        z, mu, log_var = self.encoder(x)
        reconstruction = self.decoder(z)

        # 손실 계산
        recon_loss = tf.reduce_mean(
            tf.reduce_sum(
                tf.keras.losses.binary_crossentropy(x, reconstruction), axis=(1, 2)
            )
        )
        kl_loss = -0.5 * tf.reduce_mean(
            tf.reduce_sum(
                1 + log_var - tf.square(mu) - tf.exp(log_var), axis=1
            )
        )
        total_loss = recon_loss + kl_loss

        # 지표 업데이트
        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        # 'val_loss', 'val_recon_loss' 등으로 로그가 남습니다.
        return {
            "loss": self.total_loss_tracker.result(),
            "recon_loss": self.recon_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

# --- 6. 모델 생성, 컴파일 및 학습 ---
encoder = build_encoder(latent_dim)
decoder = build_decoder(latent_dim)
vae = VAE(encoder, decoder)

# *** ⭐️ 수정된 부분 ⭐️ ***
# 'loss' 인자를 추가하여 Keras의 검증을 통과시킵니다.
# 'train_step'과 'test_step'에서 실제 손실을 계산하므로 여기서의 'mse'는 형식적인 값입니다.
vae.compile(optimizer="adam", loss="mse")

# 'validation_data'를 x_test만 전달해도 test_step이 올바르게 처리합니다.
vae.fit(x_train, epochs=10, batch_size=128, validation_data=x_test)

# --- 7. 생성 ---
z_samples = np.random.normal(size=(16, latent_dim))
x_generated = decoder.predict(z_samples)

fig, axs = plt.subplots(4, 4, figsize=(4, 4))
for i, ax in enumerate(axs.ravel()):
    ax.imshow(x_generated[i].reshape(28, 28), cmap="gray")
    ax.axis("off")
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Input, Activation, Dense, Flatten, Reshape, \
                                    Conv2D, Conv2DTranspose, Dropout, BatchNormalization, UpSampling2D
from tensorflow.keras.models import Model
import tensorflow.keras.backend as K
from tensorflow.keras.losses import mse
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------
## 1. 하이퍼파라미터 및 데이터 준비
# ----------------------------------------------------------------------

# 상수 설정
batch_size = 64
epochs = 5000  # 학습 에포크 수 (테스트를 위해 500 정도로 줄일 수 있습니다)
dropout_rate = 0.4
batch_norm = 0.9
zdim = 100  # 잠복 공간의 차원

# MNIST 데이터 로드 및 전처리
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 픽셀 값을 [-1, 1] 구간으로 변환
x_train = x_train.astype('float32') / 255.0 * 2.0 - 1.0
x_test = x_test.astype('float32') / 255.0 * 2.0 - 1.0

# Keras 입력 형태에 맞게 차원 추가 (28, 28) -> (28, 28, 1)
x_train = x_train.reshape(len(x_train), 28, 28, 1)
x_test = x_test.reshape(len(x_test), 28, 28, 1)


# ----------------------------------------------------------------------
## 2. 모델 정의 (판별망 D와 생성망 G)
# ----------------------------------------------------------------------

### 2-1. 판별망 (Discriminator, D) 설계
discriminator_input = Input(shape=(28, 28, 1)) # (28, 28, 1) 이미지 입력
x = Conv2D(64, (5, 5), activation='relu', padding='same', strides=(2, 2))(discriminator_input)
x = Dropout(dropout_rate)(x)
x = Conv2D(64, (5, 5), activation='relu', padding='same', strides=(2, 2))(x)
x = Dropout(dropout_rate)(x)
x = Conv2D(128, (5, 5), activation='relu', padding='same', strides=(2, 2))(x)
x = Dropout(dropout_rate)(x)
x = Conv2D(128, (5, 5), activation='relu', padding='same', strides=(1, 1))(x)
x = Dropout(dropout_rate)(x)
x = Flatten()(x)
discriminator_output = Dense(1, activation='sigmoid')(x)

discriminator = Model(discriminator_input, discriminator_output, name='discriminator')

# 판별망 학습 설정
discriminator.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
print("--- Discriminator Summary ---")
discriminator.summary()


### 2-2. 생성망 (Generator, G) 설계 (수정 완료)
generator_input = Input(shape=(zdim,)) # 잠복 공간 (100) 벡터 입력
x = Dense(3136)(generator_input)
x = BatchNormalization(momentum=batch_norm)(x)
x = Activation('relu')(x)
x = Reshape((7, 7, 64))(x)
x = UpSampling2D()(x) # (14, 14, 64)

# 🚨 수정된 부분: activation='same' 제거 (40행)
x = Conv2D(128, (5, 5), padding='same')(x)
x = BatchNormalization(momentum=batch_norm)(x)
x = Activation('relu')(x)

x = UpSampling2D()(x) # (28, 28, 128)

# 🚨 수정된 부분: activation='same' 제거 (44행)
x = Conv2D(64, (5, 5), padding='same')(x)
x = BatchNormalization(momentum=batch_norm)(x)
x = Activation('relu')(x)

# 🚨 수정된 부분: activation='same' 제거 (47행)
x = Conv2D(64, (5, 5), padding='same')(x)
x = BatchNormalization(momentum=batch_norm)(x)
x = Activation('relu')(x)

x = Conv2D(1, (5, 5), activation='tanh', padding='same')(x) # 출력 (28, 28, 1)
generator_output = x

generator = Model(generator_input, generator_output, name='generator')
print("\n--- Generator Summary ---")
generator.summary()


# ----------------------------------------------------------------------
## 3. GAN 모델 및 학습 함수 정의
# ----------------------------------------------------------------------

### 3-1. 전체 GAN 모델 설정
# 생성망 학습 시 판별망의 가중치는 고정
discriminator.trainable = False
gan_input = Input(shape=(zdim,))
gan_output = discriminator(generator(gan_input))
gan = Model(gan_input, gan_output, name='gan')

# 전체 GAN 모델 컴파일
gan.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
print("\n--- GAN Summary ---")
gan.summary()


### 3-2. 판별망 학습 함수 정의 (train_discriminator)
def train_discriminator(x_train):
    # 진짜 샘플 학습
    c = np.random.randint(0, x_train.shape[0], batch_size)
    real = x_train[c]
    d_loss_real = discriminator.train_on_batch(real, np.ones((batch_size, 1)))

    # 가짜 샘플 학습
    p = np.random.normal(0, 1, (batch_size, zdim))
    fake = generator.predict(p, verbose=0) # verbose=0으로 설정하여 출력 줄임
    d_loss_fake = discriminator.train_on_batch(fake, np.zeros((batch_size, 1)))

    d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
    return d_loss


### 3-3. 생성망 학습 함수 정의 (train_generator)
def train_generator():
    p = np.random.normal(0, 1, (batch_size, zdim))
    # 레이블 1 (진짜인 것처럼 속이기 위한 목표 레이블) 붙여서 GAN 모델 학습
    g_loss = gan.train_on_batch(p, np.ones((batch_size, 1)))
    return g_loss


# ----------------------------------------------------------------------
## 4. 전체 학습 루프
# ----------------------------------------------------------------------

# 학습 상황 시각화 함수 정의
def plot_generated_images(generator, epoch, examples=20, dim=(2, 10), figsize=(20, 4)):
    # 고정된 노이즈로 20개 샘플 생성
    noise = np.random.normal(0, 1, (examples, zdim))
    generated_images = generator.predict(noise, verbose=0) # verbose=0으로 설정하여 출력 줄임

    plt.figure(figsize=figsize)
    plt.suptitle(f'Epoch {str(epoch)}', fontsize=12)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i + 1)
        # 이미지 차원을 (28, 28, 1)에서 (28, 28)로 변환 후 출력
        plt.imshow(generated_images[i].reshape(28, 28), cmap='gray')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

print("\n--- 학습 시작 ---")

for i in range(epochs + 1):
    # 분별망 학습 호출
    d_loss = train_discriminator(x_train)

    # 생성망 학습 호출
    g_loss = train_generator()

    # 학습 도중 100에포크마다 상황 출력
    if i % 100 == 0:
        print(f"Epoch {i}/{epochs} | D Loss: {d_loss[0]:.4f} (Acc: {d_loss[1]:.4f}) | G Loss: {g_loss[0]:.4f} (Acc: {g_loss[1]:.4f})")
        # 이미지 출력 함수 호출
        plot_generated_images(generator, i)

print("--- 학습 완료 ---")


# ----------------------------------------------------------------------
## 5. 최종 결과 출력
# ----------------------------------------------------------------------

# 학습을 마친 후 50개 샘플을 생성하여 출력
imgs = generator.predict(np.random.normal(0, 1, (50, zdim)), verbose=0)
plt.figure(figsize=(10, 10))

# 50개 샘플을 5x10으로 시각화
for i in range(50):
    plt.subplot(5, 10, i + 1)
    plt.imshow(imgs[i].reshape(28, 28), cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.suptitle("Final Generated Samples (50)")
plt.show()
```
